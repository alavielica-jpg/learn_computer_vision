## **1. Import Libraries**

### OpenCV and NumPy:

**OpenCV (Open Source Computer Vision Library)** is one of the most widely used libraries for computer vision and image processing. It provides a comprehensive collection of optimized algorithms for tasks such as reading and displaying images, image transformations, filtering, feature detection, object detection, video processing, and many other computer vision applications. Although OpenCV is implemented primarily in C++, it offers Python bindings that make it easy to use while maintaining high performance.

In Python, OpenCV is commonly imported as `cv2`:

```python
import cv2
```

One of the key reasons OpenCV integrates so well with Python is its close relationship with **NumPy**. Every image loaded using `cv2.imread()` is represented as a **NumPy `ndarray`**, where each element corresponds to the intensity value of a pixel. A grayscale image is stored as a two-dimensional array `(height, width)`, while a color image is stored as a three-dimensional array `(height, width, channels)`.

Because images are NumPy arrays, OpenCV relies heavily on NumPy for storing and manipulating image data. NumPy provides efficient array operations such as indexing, slicing, reshaping, mathematical computations, and broadcasting, all of which can be applied directly to images. OpenCV focuses on implementing computer vision algorithms, while NumPy provides the underlying data structure and numerical operations that those algorithms operate on.

For this reason, NumPy is considered an essential dependency when working with OpenCV in Python. Although many OpenCV functions can be used without explicitly calling NumPy functions, understanding NumPy is fundamental to effectively processing and manipulating images in computer vision applications.


````python
import numpy as np
import cv2
````

## **2. What is a Digital Image?**

A digital image is a numerical representation of a visual scene. Instead of storing pictures directly, computers represent images as **NumPy arrays** composed of tiny elements called **pixels**.

Each pixel stores one or more numerical intensity values depending on the image type:

- **Grayscale image:** one intensity value per pixel.
- **Color image:** three intensity values corresponding to the Blue, Green, and Red (BGR) channels in OpenCV.

Since images are arrays, we can manipulate them mathematically using NumPy and OpenCV.

## **3. Reading Images**

OpenCV uses the `cv2.imread()` function to load an image from disk.

### **Image Loading Flags**

OpenCV provides several flags that determine how an image is loaded.

| Flag | Description |
|------|-------------|
| `cv2.IMREAD_COLOR` | Loads a color image (default). |
| `cv2.IMREAD_GRAYSCALE` | Loads a grayscale image. |
| `cv2.IMREAD_UNCHANGED` | Loads the image without modifying the number of channels. |

````python
graye_image = cv2.imread(r"C:\Users\DELL\Desktop\resources\ComputerVision\Computer Vision\vision-s-01\pic\v.jpg", cv2.IMREAD_GRAYSCALE)
color_image = cv2.imread(r"C:\Users\DELL\Desktop\resources\ComputerVision\Computer Vision\vision-s-01\pic\v.jpg", cv2.IMREAD_COLOR)
unchanged_image = cv2.imread(r"C:\Users\DELL\Desktop\resources\ComputerVision\Computer Vision\vision-s-01\pic\v.jpg", cv2.IMREAD_UNCHANGED)
````

````python
print(type(graye_image))
print(type(color_image))
print(type(unchanged_image))
````

## **5. Displaying Images**

The `cv2.imshow()` function opens a window to display an image.

The window remains open until a key is pressed. This behavior is controlled by `cv2.waitKey()`.

Finally, `cv2.destroyAllWindows()` closes all OpenCV windows.

````python
def read_image(title, image):
    cv2.imshow(title, image)
    cv2.waitKey()
    cv2.destroyAllWindows()

titles = ["Gray Image", "Color Image", "Unchanged Image"]
images = [graye_image, color_image, unchanged_image]

for title, image in zip(titles, images):
    read_image(title, image)
````

### Understanding `cv2.waitKey()`

`cv2.waitKey(delay)` waits for a keyboard event.

- `0` waits indefinitely.
- Any positive value waits that many milliseconds.

The function returns the ASCII code of the pressed key, making it useful for keyboard interactions.

## **6. Properties of Images**
### Shape

The `shape` attribute returns the dimensions of the image.

For a color image:


````python
print("Gray Image Shape:", graye_image.shape)
print(f"Gray Image Height: {graye_image.shape[0]}\nGray Image Width: {graye_image.shape[1]}")
print("=" * 40 + "\n")

print("Color Image Shape:", color_image.shape)
print(f"Color Image Height: {color_image.shape[0]}\nColor Image Width: {color_image.shape[1]}\nColor Image Channels: {color_image.shape[2]}")
print("=" * 40 + "\n")

print("Unchanged Image Shape:", unchanged_image.shape)
print(f"Unchanged Image Height: {unchanged_image.shape[0]}\nUnchanged Image Width: {unchanged_image.shape[1]}\nUnchanged Image Channels: {unchanged_image.shape[2]}")
````

The first value represents the image height (rows), the second represents the width (columns), and the third represents the number of color channels.

### Number of Dimensions

The `ndim` attribute returns the number of dimensions in the NumPy array.

- Color image → 3 dimensions
- Grayscale image → 2 dimensions

````python
print("Gray Image ndim:", graye_image.ndim)
print("Color Image ndim:", color_image.ndim)
print("Unchanged Image ndim:", unchanged_image.ndim)
````

### Data Type

The `dtype` attribute returns the data type used to store pixel values.

Most images are stored as `uint8` (unsigned 8-bit integer), allowing each pixel value to range from **0 to 255**.

````python
print("Gray Image Pixels data type:", graye_image.dtype)
print("Color Image Pixels data type:", color_image.dtype)
print("Unchanged Image Pixels data type:", unchanged_image.dtype)
````

## **7. Saving Images**

The `cv2.imwrite()` function saves an image to disk.

````python
cv2.imwrite("grayImage.jpg", graye_image)
````

## **8. Key Takeaways**

- A digital image is represented as a NumPy array.

- OpenCV loads images using `cv2.imread()`.
- Images can be displayed using `cv2.imshow()`.
- `shape`, `size`, `ndim`, and `dtype` describe the image's properties.
