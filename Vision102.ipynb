# **Two Complementary Functions**

`cv2.split()` and `cv2.merge()` are complementary functions. `cv2.split()` separates a multi-channel image into individual single-channel images (e.g., Blue, Green, and Red), while `cv2.merge()` combines these channels back into a single multi-channel image. They are useful when each channel needs to be processed independently before reconstructing the image. For simple channel access or modification, however, NumPy indexing and slicing (e.g., `img[:, :, 2]`) are generally more efficient, as they avoid creating separate channel arrays.


```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread(r"C:\Users\DELL\Desktop\resources\ComputerVision\Computer Vision\vision-s-01\pic\color-ch.png")
```

## **Using cv2.split(image) for splitting the channels**

```python
b, g, r = cv2.split(image)
```

```python
fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=4)
axes = axes.flatten()
 
channels = [r, g, b, image[..., ::-1]]
titles= ["Red Channel", "Green Channel", "Blue Channel", "Original"]

for i, (channel, title) in enumerate(zip(channels, titles)):
    axes[i].imshow(channel, cmap="gray")
    axes[i].set_title(title)
plt.tight_layout()
plt.show()
```

## **Using cv2.merge() for merging the channels**

```python
image2 = cv2.imread(r"C:\Users\DELL\Desktop\resources\ComputerVision\Computer Vision\vision-s-01\pic\x.jpg")
b, g, r = cv2.split(image2)
merged_image = cv2.merge([r, g, b])
```

```python
fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=5)
axes = axes.flatten()
 
channels = [r, g, b, image2[..., ::-1], merged_image]
titles= ["Red Channel", "Green Channel", "Blue Channel", "Original", "Merged"]

for i, (channel, title) in enumerate(zip(channels, titles)):
    axes[i].imshow(channel, cmap="gray")
    axes[i].set_title(title)
plt.tight_layout()
plt.show()
```

```python
blank_image = np.zeros_like(r, "uint8")

red = cv2.merge([blank_image, blank_image, r])
blue = cv2.merge([b, blank_image, blank_image])
green = cv2.merge([blank_image, g, blank_image])

fig, axes = plt.subplots(figsize=(20, 5), nrows=1, ncols=5)
axes = axes.flatten()
 
channels = [red, green, blue, image2[..., ::-1], merged_image]
titles= ["Red Channel", "Green Channel", "Blue Channel", "Original", "Merged"]

for i, (channel, title) in enumerate(zip(channels, titles)):
    axes[i].imshow(channel)
    axes[i].set_title(title)
plt.tight_layout()
plt.show()
```
